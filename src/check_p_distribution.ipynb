# Checking p-values

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from IPython.display import display, HTML 
from lr.training.util import get_ternary_label, filter_df_by_label
from lr.analysis.util import get_boot_columns
from lr.stats.h_testing import get_boot_p_value
```

## Checking p-values distribution for xgb - snli - sin p and h

```python
paths = ["results/snli/xgb/sin_p_h/rho_0p00_random_state_42_343.csv",
         "results/snli/xgb/sin_p_h/rho_0p25_random_state_413_13.csv",
         "results/snli/xgb/sin_p_h/rho_0p50_random_state_19_22.csv",
         "results/snli/xgb/sin_p_h/rho_0p75_random_state_322_5.csv",
         "results/snli/xgb/sin_p_h/rho_1p00_random_state_1_76.csv"]

number_of_tests = 2

for path in paths:

    for i in range(number_of_tests):

        df = pd.read_csv(path)
        boots_c = get_boot_columns(df)
        boots = df[boots_c].loc[i]

        all_p = []

        for name in boots_c:
            obs = boots.loc[name]
            new_boots = boots.drop(name)
            all_p.append(get_boot_p_value(new_boots,obs))

        all_p = pd.Series(all_p)
        all_p.hist();
        plt.show()
```

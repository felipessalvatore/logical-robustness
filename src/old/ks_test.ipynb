```python
import pandas as pd
import seaborn as sns
from lr.stats.ks import get_ks_diff
from lr.util import column_cropper
import matplotlib.pyplot as plt
```

```python
r1 = pd.read_csv("results/snli_lr_Tfidf_wordnet_p_h_syn_noun_min_0p25.csv")
r2 = pd.read_csv("results/snli_lr_Tfidf_wordnet_p_h_syn_noun_min_0p75.csv")
```

```python
ks1 = get_ks_diff(column_cropper(r1,"boot"))
fig, ax = plt.subplots(figsize=(18,9));
sns.heatmap(ks1, cmap='coolwarm', cbar=True, annot=True, fmt='.3f');
```

```python
ks2 = get_ks_diff(column_cropper(r2,"boot"))
fig, ax = plt.subplots(figsize=(18,9));
sns.heatmap(ks2, cmap='coolwarm', cbar=True, annot=True, fmt='.3f');
```
